# POLISHED CODE

```python
import requests
from bs4 import BeautifulSoup

webscraped_data = []
for i in range(0, 15, 1):
    try:
        base_url = 'https://ateneo.edu'
        extension = f'/ls/jgsom/qmit/faculty?title_op=contains&title=&page={i}'
        url = f'{base_url}{extension}'
        response = requests.get(url)
        page_text = response.text
        soup = BeautifulSoup(page_text)
        # Grab the elements which contain the data we want
        selector = '#block-views-profile-block > div > div > div > div.view-content'
        view_content_element = soup.select(selector)[0]
        selector = 'tbody > tr'
        table_rows = view_content_element.select(selector)
        # Processing table rows
        page_data = []
        for tr in table_rows:
            name_selector = '#profile-body-title'
            email_selector = '#profile-body-contact > a'
            name_element = tr.select(name_selector)
            email_element = tr.select(email_selector)
            name = ''
            email = ''
            try:
                name = name_element[0].text
            except:
                pass
            try:
                email = email_element[0].text
            except:
                pass
            page_data.append({
                'name': name,
                'email': email
            })
        webscraped_data.extend(page_data)
    except:
        print(f'Stopped at {i}')
        break
    
webscraped_data
    
```

```python
import json

with open('./webscraping-demo.json', 'w') as f:
    json.dump(webscraped_data, f, indent=4)
```

# RAW CODE

```python
import requests
from bs4 import BeautifulSoup

def scrape_page_n(url):
    response_text = requests.get(url).text
    soup = BeautifulSoup(response_text)
    
    selector = '#block-views-profile-block > div > div > div > div.view-content > table > tbody > tr'
    table_rows = soup.select(selector)
    
    contact_data = []
    for r in table_rows:
        # Get name
        selector = '#profile-body-title'
        elements = r.select(selector)
        name = elements[0].text
        # Get contact details
        email = ''
        selector = '#profile-body-contact > a'
        try:
            elements = r.select(selector)
            email = elements[0].text
        except:
            pass
        # Add to contact data
        record = {
            'name': name,
            'email': email
        }
        contact_data.append(record)

    return contact_data

data = []
for i in range(0, 15, 1):
    try:
        base_url = 'https://ateneo.edu'
        extension = f'/ls/jgsom/qmit/faculty?title_op=contains&title=&page={i}'
        url = f'{base_url}{extension}'
        data_batch = scrape_page_n(url)
        data.extend(data_batch)
    except:
        print('Done')
        break

data
```

```python
import json

filepath = './data.json'

with open(filepath, 'w') as f:
    json.dump(data, f, indent=4)
```
