# POLISHED CODE

```python
import pandas as pd

data_path = './gdp-unprocessed.csv'

df = pd.read_csv(data_path)

df = df.head(10)

def convert_gdp(x):
    charset = [
        *[str(i) for i in range(10)],
        '.'
    ]
    x = ''.join([i for i in x if i in charset])
    x = float(x) * 10 ** 12
    return x

df['Nominal GDP'] = df['GDP (abbrev.)'].apply(convert_gdp)

df = df[['Country', 'Nominal GDP']]

ax = df.plot.bar(x='Country', y='Nominal GDP', rot=45)

ax
```

# RAW CODE

```python
import pandas as pd

odf = pd.read_csv('./gdp-unprocessed.csv', index_col=0)
ndf = odf.copy()

# Convert GDP (abbrev.) to GDP
def convert_abbrev(val):
    num_chars = [*[str(i) for i in range(10)], '.']
    num_val = ''.join([i for i in val if i in num_chars])
    multiplier = val.split()[-1]
    multipliers = {
        'trillion': 10 ** 12,
        'billion': 10 ** 9,
        'million': 10 ** 6
    }
    return float(num_val) * multipliers[multiplier]

ndf['Nominal GDP'] = ndf['GDP (abbrev.)'].apply(convert_abbrev)

ndf = ndf.head(10)

ndf = ndf[['Country', 'Nominal GDP']]

ax = ndf.plot.bar(x='Country', y='Nominal GDP', rot=45)

ax
```
